### Customizing Seaborn Plots

#### set_style() method is used to set the aesthetic of the plot. It means it affects things like the color of the axes, whether the grid is active or not, or other aesthetic elements. There are five themes available in Seaborn.

 

darkgrid

whitegrid

dark

white

ticks

##### Syntax:

 

set_style(style=None, rc=None)



```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# changing the theme to dark
sns.set_style("dark")
plt.show()
```

#### Removal of Spines
Spines are the lines noting the data boundaries and connecting the axis tick marks. It can be removed using the despine() method.

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Removing the spines
sns.despine()
plt.show()
 
 

```

### Scaling the plots
It can be done using the set_context() method.

It allows us to override default parameters. 

This affects things like the size of the labels, lines, and other elements of the plot, but not the overall style. The base context is “notebook”, and the other contexts are “paper”, “talk”, and “poster”. font_scale sets the font size.

 

Syntax:

set_context(context=None, font_scale=1, rc=None)

```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 
  
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 
  
# Setting the scale of the plot
sns.set_context("talk")
  
plt.show()
```

#### Multiple plots with Seaborn
In matplotlib we use subplot 

in seaborn we use FacetGrid()

FacetGrid object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.


FacetGrid class helps in visualizing distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels.


A FacetGrid can be drawn with up to three dimensions ? row, col, and hue. The first two have obvious correspondence with the resulting array of axes; think of the hue variable as a third dimension along a depth axis, where different levels are plotted with different colors.
 

Syntax:

 

seaborn.FacetGrid( data, \*\*kwargs)

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
  
# loading dataset
data = sns.load_dataset("iris")
  
plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")
  
plt.show()
```

### Categorical data plotting

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
sns.set(style = 'darkgrid')
tips = sns.load_dataset('tips')
tips.head()
```

##catplot() : plot categorical column in the scatter form 

```python
# plot for categorical column "day" on x axis  and total bill on y-axis
sns.catplot(x="day",y="total_bill",data=tips)
```

## from this we come to knnow mostly people are paying bill on sat and sun a
### on specific day we can come to know  people mostly pays  10 to 20 dollars on thursday 
although it is somwhat same as barplot but it generates avg bill for each day and we can cant come to know how much bill has been mostly payed on each day , so through scatter points we 

```python
## adding jitter : points would come on one line  : default : True
sns.catplot(x="day",y="total_bill",data=tips,jitter=False) 
```

```python
## default : kind = strips 
## adding swarm in kind in catplot
sns.catplot(x="day",y="tip",data=tips,kind="swarm")
```

```python
## with respect to above ouput :swarm tells the frequency as well like how many people has given 2 dollar tip on thursday \
##we simply count the points corresponding to 2 tip
```

```python
sns.catplot(x="time",y="tip",data=tips)  # be default : order is lunch then dinner for time column
```

```python
# customise the order of time category
sns.catplot(x="time",y="tip",data=tips,order=["Dinner","Lunch"])
```
